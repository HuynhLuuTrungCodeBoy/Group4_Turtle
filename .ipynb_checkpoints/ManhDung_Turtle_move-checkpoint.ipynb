# Cách câu lệnh điều khiển chuyển động trong Turtle

## Di chuyển Turle

### Có 4 hướng mà một con rùa có thể di chuyển:
- Forward : phía trước
- Backward : phía sau
- Left : bên trái
- Right : bên phải

```python
import turtle
t = turtle.Turtle()

#Ví dụ vẽ hình vuông:
for i in range(4):
    t.forward(50)           #Con rùa tiến về phía trước 50 pixel
    t.right(90)             #Con rùa xoay sang phải 1 góc 90 độ
turtle.done()
```

### Ví dụ vẽ hình tròn

```python
t = turtle.Turtle()

t.circle(50)               #hình tròn có bán kính 50 pixel
turtle.done()
```

### Bạn cũng có thể sử dụng các phiên bản rút gọn của các lệnh di chuyển
- t.rt() rút gọn của t.right()
- t.fd() rút gọn của t.forward()
- t.lt() rút gọn của t.left()
- t.bk() rút gọn của t.backward()d()

### Ví dụ vẽ hình ngôi sao

```python
t = turtle.Turtle()

for i in range(5):
    t.fd(50)                #Con rùa tiến về phía trước 50 pixel 
    t.rt(144)               #Con rùa xoay sang phải 1 góc 144 độ
turtle.done()
```

### Màn hình được chia thành 4 góc phần tư như bên dưới: 
- Điều này sẽ hỗ trợ các bạn di chuyển đến bất cứ vị trí nào các bạn muốn trên màn hình

- Màn hình được chia làm 4 góc phần tư. Điểm mà con rùa được định vị ban đầu là gốc tọa độ (0,0)
- Để di chuyển con rùa đến bất kì khu vực nào khác trên màn hình, chúng ta sử dụng câu lệnh goto(a,b) và nhập tọa độ tương ứng

```python
t.goto(100,100)
```

### Để đưa con rùa về vị trí ban đầu, ta sử dụng câu lệnh home()

```python
t.home()
```

### Trong thư viện turle của Python, câu lệnh penup() và pendown() sẽ được sử dụng để kiểm soát việc vẽ của con rùa
- Câu lệnh pepup() : con rùa vẫn di chuyển theo các câu lệnh nhưng không để lại dấu vết trên màn hình
- Câu lệnh pepdown() : ngược lại với penup(), con rùa sẽ bắt đầu vẽ lại. Khi con rùa di chuyển sau câu lệnh pepdown(), nó sẽ để lại dấu vết trên màn hình

```python
t = turtle.Turtle()

# Câu lệnh gọi con trỏ tại tọa độ bất kì nào đó:
t.penup()
t.goto(-250, 100)
t.pendown()

#Vẽ chữ 'T'
t.forward(200)
t.back(100)
t.right(90)
t.forward(200)

#Chuyển sang vẽ chữ tiếp theo
t.penup()
t.back(200)
t.left(90)
t.forward(150)
t.pendown()

#Vẽ chữ 'I'
t.right(90)
t.forward(200)

#Chuyển sang vẽ chữ tiếp theo
t.penup()
t.back(200)
t.left(90)
t.forward(50)
t.pendown()

#Vẽ chữ 'M'
t.right(90)
t.forward(200)
t.back(200)
t.left(30)
t.forward(100)
t.left(120)
t.forward(100)
t.right(150)
t.forward(200)
turtle.done()
```
